Год рождения

Имеется словарь, в котором хранятся имена и возраст работников некой компании. Необходимо узнать год рождения работника. Напишите программу которая сможет определить дату рождения работника, имя которого ввели. Принять, что у каждого работника уже был день рождения в этом году.

Формат ввода:
число n - количество работников в компании
словарь в формате имя:возраст
имя работника, дату рождения которого надо узнать

Формат вывода:
Год рождения этого работника

```python
dict1 = {}
n = int(input('введите число работников: '))
for i in range(n):
    a = input('введите имя сотрудника: ')
    b = int(input('введите возраст сотрудника: '))
    dict1[a] = b
c = input('введите имя работника, год рождения которого хотите узнать: ')
d = int(dict1[c])
e = 2025 - d
print(e)
```

Пете в школе задали сделать доклад на свободную тему. Он уже придумал несколько, но чтобы не было так, что он случайно выбрал тему, которая уже
у кого-то есть, он решил свериться с одноклассниками. Если какая-то из тем повторяется, то Петя уберёт её из списка идей.

Формат ввода:
число n количества тем, выбранных Петей
выписать темы, которые выбрал Петя
число m количества тем, выбранных его одноклассником
выписать темы, которые выбрал Петин одноклассник

Формат вывода:
темы, которые нужно убрать

```python
n = int(input('число n количества тем, выбранных Петей: '))
set1 = set({})
set2 = set({})
for i in range(n):
    set1.add(input('темы, которые выбрал Петя: '))
m = int(input('число m количества тем, выбранных его одноклассником: '))
for j in range(m):
    set2.add(input('выписать темы, которые выбрал Петин одноклассник: '))
a = set1 & set2
print(*a)
```
